```python
import pandas as pd
import ast
import re
```

```python
movies = pd.read_csv("../data/tmdb_5000_movies.csv")
credits = pd.read_csv("../data/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies.info()
```

```python
df = movies.merge(
    credits,
    left_on = 'id',
    right_on = 'movie_id',
    how = 'left',
)
```

```python
df = df[['original_title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.fillna('',inplace=True)
```

```python
df.isnull().sum()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
df['overview'] = df['overview'].apply(clean_text)
```

```python
df['overview'].head()
```

```python
def parse_genres(genres):
    genres = ast.literal_eval(genres)
    return ' '.join([g['name'] for g in genres])
```

```python
df['genres'] = df['genres'].apply(parse_genres)
```

```python
df['genres'].head()
```

```python
def parse_keywords(keywords):
    keywords = ast.literal_eval(keywords)
    return ' '.join([k['name'] for k in keywords])
```

```python
df['keywords'] = df['keywords'].apply(parse_keywords)
```

```python
df['keywords'].head()
```

```python
def parse_cast(cast):
    cast = ast.literal_eval(cast)
    return ' '.join([c['name'] for c in cast[:5]])
```

```python
df['cast'] = df['cast'].apply(parse_cast)
```

```python
df['cast'].head()
```

```python
def get_director(crew):
    crew = ast.literal_eval(crew)
    for member in crew:
        if member['job'] == 'Director':
            return member['name']
    return ''
```

```python
df['director'] = df['crew'].apply(get_director)
```

```python
df['director'].head()
```

```python
df.drop(columns=['crew'] , inplace = True)
```

```python
for col in ['genres','keywords','cast','director']:
    df[col] = df[col].apply(clean_text)
```

```python
df['content'] = (
    df['overview'] + ' ' +
    df['genres'] + ' ' +
    df['keywords'] + ' ' +
    df['cast'] + ' ' +
    df['director']
    
)
```

```python
df.to_csv('../data/movies_cleaned.csv' , index=False)
```
